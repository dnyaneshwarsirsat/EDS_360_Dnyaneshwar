```python
 import numpy as np
 import pandas as pd
 data=pd.read_csv("/content/all_data (1).csv")
 print(data)
```

```python
# Find NAN
nan_df = data[data.isnull().any(axis=1)]
display(nan_df.head())
data = data.dropna (how='all')
data.head ( )
```

```python
#to work on wrong format
data = data[data['Order Date'].str[0:2]!="Or"]
print(data)
```

```python
#make columns correct type
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])
data['Price Each'] = pd.to_numeric(data['Price Each'])
print(data)
```

```python
#add differnt colums #month
data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int32')
data.head()
```

```python
#month using datatime
import datetime as dt
data['Month 2'] = pd.to_datetime(data['Order Date']).dt.month
data.head()
```

```python
#getting city name ( address)
def get_city(address):
    return address.split(",")[1].strip("")

def get_state(address):
    return address.split(",")[2].strip("")[1]

data['City'] = data ['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
data.head()
```

```python
#add sales
data['Sales'] = data['Quantity Ordered'].astype('int') * data['Price Each'].astype('float')
data.groupby(['Month']).sum()
```
